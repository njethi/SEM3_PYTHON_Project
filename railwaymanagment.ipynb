```python
class Railway:
    def __init__(self):
        self.train={"Ahmedabad To Mumbai":["Duranto","Shatabdi","GujaratMail","RanakpurExpress","SaurashtraMail"] , "Mumbai To Ahmedabad":["Duranto","Shatabdi","GujaratMail","RanakpurExpress","SaurashtraMail"] , "Ahmedabad To Pune":["Duranto","Ahinsa"] , "Pune To Ahmedabad":["Duranto","Ahinsa"] , "Ahmedabad To Delhi":["Ashram","Rajdhani"] , "Delhi To Ahmedabad":["Ashram","Rajdhani"]}
        self.depTime={"Ahmedabad To Mumbai":["23:45","15:10","22:00","23:05","23:45"],"Mumbai To Ahmedabad":["23:25","06:25","22:05","23:00","22:45"],"Ahmedabad To Pune":["22:30","20:10"],"Pune To Ahmedabad":["21:35","19:40"],"Ahmedabad To Delhi":["18:50","17:45"],"Delhi To Ahmedabad":["15:40","19:55"]}
        self.arrTime={"Ahmedabad To Mumbai":["06:00","21:40","06:25","07:10","08:10"],"Mumbai To Ahmedabad":["05:50","12:45","06:35","07:05","06:50"],"Ahmedabad To Pune":["07:10","07:35"],"Pune To Ahmedabad":["06:15","06:50"],"Ahmedabad To Delhi":["10:10","07:30"],"Delhi To Ahmedabad":["07:40","09:40"]}
        self.fac=0
        self.sac=0
        self.tac=0
        self.cc=0
        self.ec=0
        self.adi=0
        self.mmct=0
        self.ndls=0
        self.pn=0
        self.name=""
        self.fromStation=""
        self.toStation=""
        self.select_class=""
        self.coach_number=""
        self.select_train=""
        self.sdt=""
        self.sat=""
        self.meal=""
        self.fare=0
        self.date=0
        self.month=0
        self.year=2024
        self.doj=""
        self.select_seat=0
        self.ename=""
        self.edeg=""
        self.edept=""
        self.n=[]
        self.f=[]
    def book_train(self,name,fst,tst):
        self.name=name
        self.fromStation=fst
        self.toStation=tst
        self.n.append(self.name)
        if self.fromStation=="Ahmedabad":
            self.adi+=1
        elif self.fromStation=="Mumbai":
            self.mmct+=1
        elif self.fromStation=="Pune":
            self.pn+=1
        elif self.fromStation=="Delhi":
            self.ndls+=1
    def train_selection(self):
        c=self.fromStation+" To "+self.toStation
        e=self.train[c]
        f=self.depTime[c]
        g=self.arrTime[c]
        flag=True
        while flag:
            for i in range(len(e)):
                print("Press",(i+1),"For",e[i],"Departure Time =",f[i],"Arival Time =",g[i])
            ip=int(input("Enter Your Choice "))
            if((ip>0)and(ip<=len(e))):
                self.select_train=e[ip-1]
                self.sdt=f[ip-1]
                self.sat=g[ip-1]
                flag=False
            else:
                print("Enter Valid Choice")
    def date_of_journey(self):
        self.date_flag=True
        while self.date_flag:
            self.date=int(input("Enter Date "))
            self.month=int(input("Enter Month "))
            if((self.date>=1)and(self.date<=30)and(self.month>=1)and(self.month<=12)):
                self.doj=str(self.date)+"/"+str(self.month)+"/"+str(self.year)
                self.date_flag=False
            else:
                print("Enter Valid Choice")
    def class_selection(self):
        self.class_select_flag=True
        while self.class_select_flag:
            self.class_select=int(input("Select Your Class : 1)1AC 2)2AC 3)3AC 4)CC 5)EC "))
            if self.class_select==1:
                self.select_class="1AC"
                self.fac+=1
                self.class_select_flag=False
            elif self.class_select==2:
                self.select_class="2AC"
                self.sac+=1
                self.class_select_flag=False
            elif self.class_select==3:
                self.select_class="3AC"
                self.tac+=1
                self.class_select_flag=False
            elif self.class_select==4:
                self.select_class="CC"
                self.cc+=1
                self.class_select_flag=False
            elif self.class_select==5:
                self.select_class="EC"
                self.ec+=1
                self.class_select_flag=False
            else:
                print("Enter Valid Choice")
    def coach_and_seat_selection(self):
        if self.select_class=="1AC":
            self.category="H"
            self.allocateNumber=str(1)
            self.coach_number=self.category+self.allocateNumber
            self.seat_select_flag=True
            while self.seat_select_flag:
                self.select_seat=int(input("Enter Seat "))
                if((self.select_seat>0)and(self.select_seat<25)):
                    self.seat_select_flag=False
                else:
                    print("Enter Valid Choice")
        elif self.select_class=="2AC":
            self.category="A"
            self.allocateNumber=input("Enter Allocate Number ")
            self.coach_number=self.category+self.allocateNumber
            self.seat_select_flag=True
            while self.seat_select_flag:
                self.select_seat=int(input("Enter Seat "))
                if((self.select_seat>0)and(self.select_seat<53)):
                    self.seat_select_flag=False
                else:
                    print("Enter Valid Choice")
        elif self.select_class=="3AC":
            self.category="B"
            self.allocateNumber=input("Enter Allocate Number ")
            self.coach_number=self.category+self.allocateNumber
            self.seat_select_flag=True
            while self.seat_select_flag:
                self.select_seat=int(input("Enter Seat "))
                if((self.select_seat>0)and(self.select_seat<73)):
                    self.seat_select_flag=False
                else:
                    print("Enter Valid Choice")
        elif self.select_class=="CC":
            self.category="C"
            self.allocateNumber=input("Enter Allocate Number ")
            self.coach_number=self.category+self.allocateNumber
            self.seat_select_flag=True
            while self.seat_select_flag:
                self.select_seat=int(input("Enter Seat "))
                if((self.select_seat>0)and(self.select_seat<79)):
                    self.seat_select_flag=False
                else:
                    print("Enter Valid Choice")
        elif self.select_class=="EC":
            self.category="E"
            self.allocateNumber=input("Enter Allocate Number ")
            self.coach_number=self.category+self.allocateNumber
            self.seat_select_flag=True
            while self.seat_select_flag:
                self.select_seat=int(input("Enter Seat "))
                if((self.select_seat>0)and(self.select_seat<57)):
                    self.seat_select_flag=False
                else:
                    print("Enter Valid Choice")
    def int_fare(self):
        if((self.fromStation=="Ahmedabad")and(self.toStation=="Mumbai")):
            if(self.select_class=="1AC"):
                self.fare=2100
            elif(self.select_class=="2AC"):
                self.fare=1600
            elif(self.select_class=="3AC"):
                self.fare=1100
            elif(self.select_class=="CC"):
                self.fare=1200
            elif(self.select_class=="EC"):
                self.fare=1800
        elif((self.fromStation=="Mumbai")and(self.toStation=="Ahmedabad")):
            if(self.select_class=="1AC"):
                self.fare=2100
            elif(self.select_class=="2AC"):
                self.fare=1600
            elif(self.select_class=="3AC"):
                self.fare=1100
            elif(self.select_class=="CC"):
                self.fare=1200
            elif(self.select_class=="EC"):
                self.fare=1800
        elif((self.fromStation=="Ahmedabad")and(self.toStation=="Pune")):
            if(self.select_class=="1AC"):
                self.fare=2300
            elif(self.select_class=="2AC"):
                self.fare=1800
            elif(self.select_class=="3AC"):
                self.fare=1200
            elif(self.select_class=="CC"):
                self.fare=1300
            elif(self.select_class=="EC"):
                self.fare=1900
        elif((self.fromStation=="Pune")and(self.toStation=="Ahmedabad")):
            if(self.select_class=="1AC"):
                self.fare=2300
            elif(self.select_class=="2AC"):
                self.fare=1800
            elif(self.select_class=="3AC"):
                self.fare=1200
            elif(self.select_class=="CC"):
                self.fare=1300
            elif(self.select_class=="EC"):
                self.fare=1900
        elif((self.fromStation=="Ahmedabad")and(self.toStation=="Delhi")):
            if(self.select_class=="1AC"):
                self.fare=3300
            elif(self.select_class=="2AC"):
                self.fare=2800
            elif(self.select_class=="3AC"):
                self.fare=1700
            elif(self.select_class=="CC"):
                self.fare=2300
            elif(self.select_class=="EC"):
                self.fare=2700
        elif((self.fromStation=="Delhi")and(self.toStation=="Ahmedabad")):
            if(self.select_class=="1AC"):
                self.fare=3300
            elif(self.select_class=="2AC"):
                self.fare=2800
            elif(self.select_class=="3AC"):
                self.fare=1700
            elif(self.select_class=="CC"):
                self.fare=2300
            elif(self.select_class=="EC"):
                self.fare=2700
    def meal_option(self):
        self.mflag=True
        while self.mflag:
            self.a=int(input("Enter Choice Of Meal 1.Sandwitch-60RS 2.Pizza-120RS 3.Bhel-40RS 4.Samosa-30RS 5.Puff-35RS 6.NoMeal "))
            if self.a==1:
                self.fare=self.fare+60
                self.meal="Sandwitch"
                self.mflag=False
            elif self.a==2:
                self.fare=self.fare+120
                self.meal="Pizza"
                self.mflag=False
            elif self.a==3:
                self.fare=self.fare+40
                self.meal="Bhel"
                self.mflag=False
            elif self.a==4:
                self.fare=self.fare+30
                self.meal="Samosa"
                self.mflag=False
            elif self.a==5:
                self.fare=self.fare+35
                self.meal="Puff"
                self.mflag=False
            elif self.a==6:
                self.meal="NoMeal"
                self.mflag=False
            else:
                print("Enter Valid Choice")
        print("Your Have To Pay Rupess",self.fare)
        self.f.append(self.fare)
    def print_details(self):
        print("Passengers Details :")
        print("Name =",self.name)
        print("From Station =",self.fromStation)
        print("To Station =",self.toStation)
        print("Train Name =",self.select_train)
        print("Mead Included =",self.meal)
        print("Departure Time =",self.sdt)
        print("Arival Time =",self.sat)
        print("Date Of Journey =",self.doj)
        print("Total Fare =",self.fare)
        print("Class =",self.select_class)
        print("Coach Number =",self.coach_number)
        print("Seat Number =",self.select_seat)
    def take_employee_details(self , ename , edept , edeg):
        self.ename=ename
        self.edept=edept
        self.edeg=edeg
    def print_employee_details(self):
        print("Employee Details :")
        print("Name Of Employee =",self.ename)
        print("Department Of Employee =",self.edept)
        print("Designation Of Employee =",self.edeg)
    def view_graphics(self):
        self.view_graphics=True
        while self.view_graphics:
            self.vgchoice=int(input("Enter Choice For View Graph 1)Pie Chart Of Classes 2)Bar Chart Of Departure Station 3)Exit 4)Line Chart Of Persons and Fare "))
            if self.vgchoice==1:
                import matplotlib.pyplot as plt
                import numpy as np
                self.dclas=np.array([self.fac,self.sac,self.tac,self.cc,self.ec])
                self.label=np.array(["1AC" , "2AC" , "3AC" , "CC" , "EC"])
                plt.pie(self.dclas,labels=self.label,autopct="%0.2f%%")
                plt.legend()
                plt.title("Pie Chart Of Classes")
                plt.show()
            elif self.vgchoice==2:
                import matplotlib.pyplot as plt
                import numpy as np
                self.station=np.array(["Ahmedabad","Mumbai","Pune","Delhi"])
                self.numberOfPeople=np.array([self.adi,self.mmct,self.pn,self.ndls])
                self.color=np.array(["green","blue","red","yellow"])
                plt.bar(self.station,self.numberOfPeople,color=self.color,label=self.station)
                plt.xlabel("Name Of Departure City")
                plt.ylabel("Number Of People")
                plt.title("Bar Chart Of Station")
                plt.legend()
                plt.show()
            elif self.vgchoice==3:
                self.view_graphics=False
            elif self.vgchoice==4:
                import matplotlib.pyplot as plt
                plt.plot(self.n,self.f)
                plt.title("Line Chart Of People and Fare")
                plt.xlabel("Name Of Person")
                plt.ylabel("Fare")
                plt.show()
            else:
                print("Enter Valid Choice")
railway=Railway()
flag=True
while flag:
    choice=int(input("Enter Choice 1.EmployeeLogin 2.UserLogin 3.Exit "))
    if choice==1:
        ename=input("Enter Name Of The Employee ")
        edept=input("Enter Department Of The Employee ")
        edeg=input("Enter Designation Of Employee ")
        railway.take_employee_details(ename,edept,edeg)
        railway.print_employee_details()
        railway.view_graphics()
    elif choice==2:
        name=input("Enter Name ")
        fromstation=input("Enter From Station ")
        tostation=input("Enter To Station ")
        railway.book_train(name,fromstation,tostation)
        railway.train_selection()
        railway.date_of_journey()
        railway.class_selection()
        railway.coach_and_seat_selection()
        railway.int_fare()
        railway.meal_option()
        railway.print_details()
    elif choice==3:
        flag=False
    else:
        print("Enter Valid Choice")
```
